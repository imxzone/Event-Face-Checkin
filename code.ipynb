```python
import os

# 1. In thư mục hiện tại đang chạy
print("Thư mục hiện tại:", os.getcwd())

# 2. THAY ĐƯỜNG DẪN NÀY nếu ảnh nằm trong thư mục con, ví dụ: r"D:\Images"
folder = "data/khachmoi_full"   # "." = thư mục hiện tại

# 3. Liệt kê file trong thư mục để kiểm tra
print("\nDanh sách file trong thư mục:")
for f in os.listdir(folder):
    print(" -", f)

print("\nBắt đầu rename...\n")

count = 0

for filename in os.listdir(folder):
    # ép về dạng lowercase để tránh .PNG / .Png
    lower_name = filename.lower()

    # điều kiện: bắt đầu bằng 'f' và kết thúc bằng '.png'
    if lower_name.startswith("f") and lower_name.endswith(".jpg"):
        new_name = "A" + filename[1:]  # đổi F đầu thành A, giữ nguyên phần còn lại

        old_path = os.path.join(folder, filename)
        new_path = os.path.join(folder, new_name)

        print(f"Đổi: {filename} -> {new_name}")
        os.rename(old_path, new_path)
        count += 1

print(f"\nHoàn tất! Đã đổi {count} file.")
```
